```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
N = 10
print(iris.feature_names)
X = iris.data[:N, :2]

plt.figure(figsize=(10,10))
plt.plot(X.T, 'o:')
plt.xticks(range(2), ["sepal length", "sepal width"])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title("  ")
plt.legend([" {}".format(i + 1) for i in range(N)])
plt.show()
```

```python
import seaborn as sns
import pandas as pd

plt.figure(figsize=(8, 8))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color=".2", marker="s")
for i in range(N):
    ax.text(X[i, 0] - 0.05, X[i, 1] + 0.03, " {}".format(i + 1))
plt.xlabel(" ")
plt.ylabel(" ")
plt.title("   (2 )")
plt.axis("equal")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
X_low = pca.fit_transform(X)

X2 = pca.inverse_transform(X_low)

plt.figure(figsize=(7, 7))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color=".2", marker="s")
for i in range(N):
    d = 0.03 if X[i, 1] > X2[i, 1] else -0.04
    ax.text(X[i, 0] - 0.065, X[i, 1] + d, " {}".format(i + 1))
    plt.plot([X[i, 0], X2[i, 0]], [X[i, 1], X2[i, 1]], "k--")
plt.plot(X2[:, 0], X2[:, 1], "o-", markersize=10)
plt.plot(X[:, 0].mean(), X[:, 1].mean(), markersize=10, marker="D")
plt.axvline(X[:, 0].mean(), c='r')
plt.axhline(X[:, 1].mean(), c='r')
plt.grid(False)
plt.xlabel(" ")
plt.ylabel(" ")
plt.title("Iris  1 ")
plt.axis("equal")
plt.show()
```
